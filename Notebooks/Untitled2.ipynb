```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import missingno as msno
```

```python
df = pd.read_csv("BostonHousing.csv")
```

```python
df.head()
```

```python
ax = df.groupby('CHAS').mean().MEDV.plot(kind = "bar", figsize = [5, 3])

ax.set_label("Avg. MEDV")

plt.tight_layout()

plt.show()

```

```python
ax = df["MEDV"].hist()
ax.set_xlabel("MEDV")
ax.set_ylabel('count')
plt.show()
```

```python
ax = df.boxplot(column='MEDV', by='CHAS')
ax.set_ylabel('MEDV')
plt.suptitle('') # Suppress the titles
plt.title('')
plt.show()
ax = df.boxplot(column='MEDV', by='RAD')
ax.set_ylabel('MEDV')
plt.suptitle('') # Suppress the titles
plt.title('')
plt.show()
```

```python
df.corr().round(3)
```

```python
corr = df.corr() # create correlation table
type(corr)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11, 7)
sns.heatmap(corr, annot=True, fmt=".1f", cmap="RdBu", center=0, ax=ax)
plt.show()
```
